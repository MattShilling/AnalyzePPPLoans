```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib notebook
```

```python
df = pd.read_csv('foia_150k_plus.csv')
```

```python
eths = df.loc[df['RaceEthnicity'] != 'Unanswered']['RaceEthnicity']
eths_list = eths.unique().tolist()
eths_counts = []

for eth in eths_list:
    count = sum(eths == eth)
    eths_counts.append(count)

print(eths_list)
print(eths_counts)
```

```python
def autolabel(bars):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{:.2f}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')
```

```python
fig, ax = plt.subplots()
x = np.arange(len(eths_list))
bars = ax.bar(x, height=eths_counts)
ax.set_title("Race/Ethnicity vs Loans >= %150")
ax.set_xticks(x)
ax.set_xticklabels(eths_list, rotation=80)

autolabel(bars)
fig.tight_layout()
plt.rcParams["figure.figsize"] = (8,8)
```
